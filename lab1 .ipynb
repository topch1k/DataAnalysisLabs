```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
num_points = 100

A = -5
B = 3
x = np.random.random(size=num_points)
noise = np.random.random(size=num_points)

y = A*x+B+noise

data = {'X': x, 'Y': y}
df = pd.DataFrame(data, columns = ['X','Y'])
X = df['X'].values
Y = df['Y'].values

D_a = 0
D_b = 0

L = 0.1

n = float(len(X))
a = -5
b = 3

for i in range(2*len(X)):
    Y_pred = a*X + b
    D_a = (-2/n) * sum(X * (Y - Y_pred))
    D_b = (-2/n) * sum(Y - Y_pred)
    a = a - L * D_a
    b = b - L * D_b

Y_pred = a*X + b
print (f'a_predicted = {a} \na = {A} \nb_predicted = {b} \nb = {B}')

plt.show()

plt.scatter(X, Y, color = 'purple')
plt.plot(X, Y_pred, color='yellow')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Logistic Regression')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

num_points = 200
a0=np.array([0,0])
b0=0
learning_rate_a=0.1
learning_rate_b=0.1
mean_1 = [6, 6]
cov_1 = [[3, 1], [1, 3]]
x_1 = np.random.multivariate_normal(mean_1, cov_1, num_points)
mean_2 = [4, 4]
cov_2 = [[3, 1], [1, 3]]
x_2 = np.random.multivariate_normal(mean_2, cov_2, num_points)

plt.title('Mixed points')
plt.scatter(x_1[:, 0], x_1[:, 1], marker = '*', color='purple')
plt.scatter(x_2[:, 0], x_2[:, 1], marker = '^', color='yellow')
plt.grid()


def sigmoid(value):
    return 1 / (1 + np.exp(-value))

def gradient_descent(x, y, iterations, learning_rate_a, learning_rate_b):
    a = a0
    b = b0
    for i in range(iterations):
        f = sigmoid(x.dot(a) + b)
        da = np.dot(x.T, (f - y))/x.shape[0]
        db = (f - y).mean()
        a = a - learning_rate_a * da
        b = b - learning_rate_b * db
    return a, b

def predict(x, a, b):
    predict = np.zeros(x.shape[0])
    f = sigmoid(x.dot(a) + b)
    for i in range(x.shape[0]):
        if(f[i] >= 0.5):
            predict[i] = 1
    return predict


iterations = 1000

x = np.concatenate([x_1,x_2])
y = np.hstack((np.array([0] * num_points), np.array([1]*num_points)))

a, b = gradient_descent(x, y, iterations,  learning_rate_a, learning_rate_b)
result = predict(x,a,b)
print("Accuracy: ", 1 - abs(y-result).mean())

plt.figure(figsize=(6,4))
plt.plot(x[:,0][result==0], x[:,1][result==0], 'r.', marker = '*', color='purple')
plt.title('Separated points')
plt.plot(x[:,0][result==1], x[:,1][result==1], 'b.', marker = '^', color='yellow')
plt.grid()
plt.show()
```
